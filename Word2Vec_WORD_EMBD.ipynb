```python
import pandas as pd
import gensim
```

```python
df=pd.read_csv('/content/Womens Clothing E-Commerce Reviews.csv')#load data

```

```python
df.head()
```

```python
df.shape
```

```python
df[df['Review Text']=='']
```

```python
#have some blank values in review txt fill with na
```

```python
df['Review Text'] = df['Review Text'].fillna('')
```

 Preprocessing & Tokenization

```python
rv_tx=df['Review Text'].apply(gensim.utils.simple_preprocess)
```

```python
rv_tx.loc[5]
```

```python
#model building
```

```python
model=gensim.models.Word2Vec(
    window=10,
    min_count=2,
    workers=4,
)
```

```python
#Build Vocabulary
```

```python
model.build_vocab(rv_tx,progress_per=1000)
```

```python
#train model
```

```python
model.corpus_count#he total number of training examples
```

```python
model.epochs#defult epochs
```

```python
model.train(rv_tx,total_examples=model.corpus_count,epochs=model.epochs)
```

```python
#find similar word
```

```python
model.wv.most_similar("bad")
```

```python
model.wv.most_similar('good')
```

```python
model.wv.similarity(w1='good',w2='great')
```
